```python
import os
import numpy as np
import pandas as pd
import sys

def convert_data(df):
    df['weather'] = df.attributes.apply(lambda x: x['weather'])
    df['timeofday'] = df.attributes.apply(lambda x: x['timeofday'])
    df['scene'] = df.attributes.apply(lambda x: x['scene'])
    # Remove unneeded / redundant columns
    df.drop(columns=['timestamp', 'attributes'], inplace=True)
    # Remove duplicates (if any)
    # df.drop_duplicates(subset=['name'], keep='first', inplace=True)
    # reset index
    df.reset_index(inplace=True, drop=True)
    return df
```

```python
# paths to json file that I am including in this folder
path_newjson = "/home/till/SharedFolder/CurrentDatasets/bdd100k_sorted/annotations/bdd100k_sorted_train_A_over.json"
# paths to original BDD train and val jsons
path_train = "/home/till/data/driving/BerkeleyDeepDrive/bdd100k/labels/bdd100k_labels_images_train.json"
path_val = "/home/till/data/driving/BerkeleyDeepDrive/bdd100k/labels/bdd100k_labels_images_val.json"
```

```python
# data according to new json
data = pd.read_json(path_newjson)
data = convert_data(data)
```

```python
image_base_path = "/home/till/data/driving/BerkeleyDeepDrive/bdd100k/labels"
data["name"] = image_base_path + "/" + data["name"].astype(str)
```

```python
# data according to original BDD
data_orig = pd.read_json(path_train)
data_add = pd.read_json(path_val)
data_orig = pd.concat([data_orig, data_add], axis=0)
data_orig = convert_data(data_orig)
```

```python
# check whether the same image occurs with different weather calsses in the json to be investigated
for i in range(data.shape[0]):
    for j in range(data.shape[0]):
        if i != j:
            if data.loc[i,"name"] == data.loc[j,"name"] and data.loc[i,"weather"] != data.loc[j,"weather"]:
                print("weather mismatch error")
                print(data.loc[i, "name"], data.loc[i, "weather"])
                print(data.loc[j, "name"], data.loc[j, "weather"])
```

```python
# check whether the weather classes in the json under investigation match those in the original BDD jsons
for j in range(data.shape[0]):
    im_name = data.loc[j,"name"]
    im_weather = data.loc[j,"weather"]
    inx = np.array(data_orig.query("name == @im_name").index)
    orig_weather = data_orig.loc[inx[0], "weather"]
    orig_name = data_orig.loc[inx[0], "name"]
    if im_weather != orig_weather and \
    (set([im_weather]).isdisjoint(["cloudy", "overcast", "partly cloudy"]) and set([orig_weather]).isdisjoint(["cloudy", "overcast", "partly cloudy"])):
        print("weather mismatch error")
        print(im_name, im_weather)
        print(orig_name, orig_weather)
```
