### Aspect Ratios & Resolutions ###

```python
df_mapillary_aspectratios = pd.read_csv("mapillary_aspectratios.csv")
df_mapillary_aspectratios["aspectratios"] = (df_mapillary_aspectratios["width"] / df_mapillary_aspectratios["height"]).round(2).astype(str) + ":1"
df_mapillary_aspectratios["resolutions"] = df_mapillary_aspectratios["width"].astype(str).str.cat(df_mapillary_aspectratios[["height"]].astype(str), sep="x")
```

```python
ax_ar = df_mapillary_aspectratios[df_mapillary_aspectratios["set"] == "testing"].groupby("aspectratios")["file"].count().plot(kind = "bar", title = "Mapillary Vistas Test Set: Aspect Ratios", figsize = (16, 9))
for p in ax_ar.patches:
    ax_ar.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax_ar.set_ylabel("# Frames");
```

```python
ax_res = df_mapillary_aspectratios[df_mapillary_aspectratios["set"] == "testing"].groupby("resolutions")["file"].count().plot(kind = "bar", title = "Mapillary Vistas Test Set: Resolutions", figsize = (21, 9))
for p in ax_res.patches:
    ax_res.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax_res.set_ylabel("# Frames");
```

## Validation Set ##

```python
labelfile_valid = "mapillary-vistas-dataset_public_v1.1_validation.json"
df_mapillary_valid = pd.read_json(labelfile_valid)
df_mapillary_valid_attributes = pd.DataFrame(df_mapillary_valid['attributes'].tolist())
```

### Weather ###

```python
ax_weather = df_mapillary_valid_attributes["weather"].value_counts().plot(kind = "bar", title = "Mapillary Vistas Validation Set: Weather", figsize = (16, 9))
for p in ax_weather.patches:
    ax_weather.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax_weather.set_ylabel("# Frames");
```

### Time-of-Day ###

```python
ax_timeofday = df_mapillary_valid_attributes["timeofday"].value_counts().plot(kind = "bar", title = "Mapillary Vistas Validation Set: Time-of-Day", figsize = (16, 9))
for p in ax_timeofday.patches:
    ax_timeofday.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax_timeofday.set_ylabel("# Frames");
```

### Objects ###

```python
ax_objects = pd.DataFrame([item for sublist in df_mapillary_valid["labels"].tolist() for item in sublist])["category"].value_counts().plot(kind = "bar", title = "Mapillary Vistas Validation Set: Objects", figsize = (16, 9))
for p in ax_objects.patches:
    ax_objects.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax_objects.set_ylabel("# Labels");
```

### Aspect Ratios & Resolutions ###

```python
ax_ar = df_mapillary_aspectratios[df_mapillary_aspectratios["set"] == "validation"].groupby("aspectratios")["file"].count().plot(kind = "bar", title = "Mapillary Vistas Validation Set: Aspect Ratios", figsize = (16, 9))
for p in ax_ar.patches:
    ax_ar.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax_ar.set_ylabel("# Frames");
```

```python
ax_res = df_mapillary_aspectratios[df_mapillary_aspectratios["set"] == "validation"].groupby("resolutions")["file"].count().plot(kind = "bar", title = "Mapillary Vistas Validation Set: Resolutions", figsize = (21, 9))
for p in ax_res.patches:
    ax_res.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
ax_res.set_ylabel("# Frames");
```
